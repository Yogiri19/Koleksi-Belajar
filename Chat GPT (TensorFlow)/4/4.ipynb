```python
total = tf.Variable(0)
for i in tf.range(1, 11): # dari 1 - 10
    total.assign_add(1) # total + 1

print(f"Jumlah: \n{total.numpy()}")
```

```python
nilai = tf.constant(80)
if nilai.numpy() >= 75:
    print(f"Lulus")
else:
    print(f"Tidak lulus")
```

```python
nilai = tf.constant(80)

hasil = tf.cond( # IF ELSE versi tensorflow
    nilai >= 75,
    lambda: tf.constant("Lulus"), 
    lambda: tf.constant("Tidak Lulus")
)

print(hasil.numpy().decode())
```

```python
@tf.function # @tf_function = mengubah fungsi python menjadi Tensorflow Graph
def tambah (a,b):
    return a + b

hasil = tambah(10, 20)
print(f"Hasil: \n{hasil}")
```

```python
@tf.function
def hitung(a,b):
    tambah = a + b
    kurang = a - b
    kali = a * b
    bagi = a / b

    return tambah, kurang, kali, bagi

hasil = hitung(20.0, 5.0)
tf.print(hasil)
```

```python
@tf.function
def jumlah(n):
    total = tf.constant(0)

    for i in tf.range(1, n + 1): # (1, 11) jadinya dari 1 sampai 10
        total += i
    return total

print(jumlah(10))
```

```python
@tf.function
def cek_nilai(nilai):
    if nilai >= 75:
        return "Lulus"
    else:
        return "Tidak Lulus"
    
print(cek_nilai(tf.constant(80)))
```

```python
i = tf.constant(1)
total = tf.constant(0)

kondisi = lambda i, total: i <= 5
badan = lambda i, total: (
    i + 1,
    total + 1
)
i, total = tf.while_loop(
    kondisi,
    badan,
    [i, total]
)
print(total)
```

```python
n = tf.constant(5)

i = tf.constant(1)
hasil = tf.constant(1)

kondisi = lambda i, hasil: i <= n
badan = lambda i, hasil: (
    i + 1,
    hasil * i
)

i, hasil = tf.while_loop(
    kondisi, 
    badan,
    [i, hasil]
)
print(f"Faktorial: \n{hasil}")
```

```python
for i in tf.range(1,6):
    for j in tf.range(1, 6):
        hasil = i * j
        print(
            f"{i.numpy()} x {j.numpy()} = {hasil.numpy()}"
        )
```

```python
nilai = tf.constant(
    [80, 70, 90, 85, 95]
)
jumlah = tf.constant(0)
for i in nilai:
    jumlah += i

rata_rata = (
    tf.cast(jumlah, tf.float32) / tf.cast(tf.size(nilai), tf.float32)
)

print(f"Jumlah: \n{jumlah}")
print(f"Rata-Rata: \n{rata_rata}")
```
