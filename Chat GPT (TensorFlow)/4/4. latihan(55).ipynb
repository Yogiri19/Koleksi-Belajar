```python
import tensorflow as tf
print(f"versi: \n{tf.__version__}")
```

```python
# Soal Pertama
angka1 = tf.range(8)
angka2 = tf.range(
    start = 5,
    limit = 20,
    delta = 3
)

print(angka1.numpy())
print(angka2.numpy())
```

```python
# Soal Kedua
total = tf.Variable(0)

for i in tf.range(1,6):
    total.assign_add(i)
print(total.numpy())
```

```python
# Soal Ketiga
nilai = tf.constant(60)

hasil = tf.cond(
    nilai >= 75,
    lambda: tf.constat("Lulus"),
    lambda: tf.constant("Tidak Lulus")
)
print(hasil.numpy().decode())

```

```python
# Soal Keempat
@tf.function
def hitung(a,b):
    tambah = a + b
    kurang = a - b
    kali = a * b
    return tambah, kurang, kali, hasil

hasil = hitung(12.0, 3.0)
tf.print(hasil)
```

```python
# Soal Kelima
@tf.function
def jumlah(n):
    total = tf.constant(0)
    for i in tf.range(1, n + 1):
        total += i
    return total

hasil = jumlah(5)
print(hasil.numpy())


wh = tf.while_loop(
    total = tf.constant(0),
    for i in tf.range(1, hasil)


)
print(wh)
```
