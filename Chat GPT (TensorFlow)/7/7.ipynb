```python
import tensorflow as tf
print(f"Versi: {tf.__version__}")
```

Sequential = input -> layer 1 -> Layer 2 -> Output
optimizer="sgd" = Stohastic Gradient Descent -> sama seperti assign_sub()
activation="relu" 
activation = Fungsi yang mengubah output neuron
ReLU = f(x)=max(0,x)
Sigmoid = mengubah angka menjadi = 0 sampai 1
Dense = y = w1x1 + w2x2 + b
(shape) = (jumlah_data, jumlah_fitur)

```python
# y = 2x + 1
x_train = tf.constant([
    [1.0],
    [2.0],
    [3.0],
    [4.0]
])
y_train = tf.constant([
    [3.0],
    [5.0],
    [7.0],
    [9.0]
])
print(f"x_train: \n{x_train}")
print(f"y_train: \n{y_train}")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1) # Otomatis memiliki weight(w) dan Bias(b)
])
model.build(input_shape=(None, 1)) # 1 nya tergantung kolom di x_train
print(f"Model Summary: \n{model.summary()}")

print(f"Weight Awal: \n{model.weights[0].numpy()}")

print(f"Bias Awal: \n{model.weights[1].numpy()}")
```

```python
model.compile(
    optimizer="sgd",
    loss="mse"
)
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    verbose=0
)
print(f"Training selesai\n")


print(f"Weight Akhir: \n{model.weights[0].numpy()}")

print(f"Bias Akhir: \n{model.weights[1].numpy()}")

print(f"Loss Terakhir: \n{history.history['loss'][-1]}")
print(f"model: \n{model}")
```

```python
prediksi = model.predict(
    tf.constant([
        [5.0],
        [10.0],
        [20.0]
    ]),
    verbose=0
)
print(f"Prediksi")

for i, hasil in enumerate(prediksi):
    print(f"Data ke-{i+1}: {hasil[0]}")
```

```python
x = tf.constant([
    [-5.0],
    [0.0],
    [5.0]
])

hasil = tf.nn.sigmoid(x)
print(f"Input: \n{x.numpy()}")
print(f"Output Sigmoid: \n{hasil.numpy()}")
```

```python
x_train = tf.constant([
    [1.0, 2.0],
    [2.0, 3.0],
    [3.0, 4.0],
    [4.0, 5.0]
])
y_train = tf.constant([
    [8.0],
    [13.0],
    [18.0],
    [23.0]
])
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(
    optimizer="sgd",
    loss="mse"
)
model.fit(
    x_train,
    y_train,
    epochs=300,
    verbose=0
)
print(f"Training Selesai")
print(f"Weight:\n {model.weights[0].numpy()}")
print(f"Bias: \n{model.weights[1].numpy()}")

prediksi = model.predict(
    tf.constant([[10.0, 11.0]]),
    verbose = 0
)
print(f"Prediksi [10,11]: \n{prediksi}")
```

```python
model.save("model_pertama.keras")
print(f"Model Berhasil disimpan")

model_baru = tf.keras.models.load_model(
    "model_pertama.keras"
)
print(f"Model Berhasil Di-Load")

model_baru.summary()
hasil = model_baru.predict(
    tf.constant([[10.0, 11.0]]),
    verbose=0
)
print(f"Prediksi Setelah Load: \n{hasil}")
```
