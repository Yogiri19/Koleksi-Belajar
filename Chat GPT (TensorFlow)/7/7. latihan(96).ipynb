```python
x_train = tf.constant([
    [1.0],
    [2.0],
    [3.0],
    [4.0]
])
y_train = tf.constant([
    [3.0],
    [5.0],
    [7.0],
    [9.0]
])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model.build(
    input_shape=(None,1)
)
model.compile(
    optimizer="sgd",
    loss="mse"
)
history=model.fit(
    x_train,
    y_train,
    epochs=50,
    verbose=0
)
print(f"History: \n{history.history['loss'][-1]}")
```

```python
# Soal Kelima
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    verbose=0
)
prediksi = model.predict([
    tf.constant([
        [5.0]
    ])
])
print(prediksi)
for i, hasil in enumerate(prediksi):
    print(f"dat ke-1{i+1}: {hasil[0]}")
```

```python
# Soal Keenam
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    verbose=0
)
prediksi = model.predict([
    tf.constant([
        [5.0],
        [10.0],
        [20.0]
    ])
])
for i,hasil in enumerate(prediksi):
    print(f"data ke-{i+1}: {hasil}")
```

```python
# Soal Kesembilan
import tensorflow as tf
x_train = tf.constant([
    [1.0, 2.0],
    [2.0, 3.0],
    [3.0, 4.0],
    [4.0, 5.0]
])
y_train = tf.constant([
    [8.0],
    [13.0],
    [18.0],
    [23.0]
])
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model.build(input_shape=(None,2))

model.compile(
    optimizer="sgd",
    loss="mse"
)
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    verbose=0
)
prediksi = model.predict(
    tf.constant([
        [10.0, 11.0]
    ]),
    verbose=0
)
print(f"Prediksi: \n{prediksi}")
```

```python
# soal kesepuluh
import tensorflow as tf
x_train = tf.constant([
    [1.0],
    [2.0],
    [3.0],
    [4.0]
])
y_train = tf.constant([
    [3.0],
    [5.0],
    [7.0],
    [9.0]
])
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model.build(input_shape=(None,1))
model.compile(
    optimizer="sgd",
    loss="mse"
)
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    verbose=0
)

model.save("model_latihan.keras")
model_baru = tf.keras.models.load_model(
    "model_latihan.keras"
)

prediksi = model_baru.predict(
    tf.constant([[10.0]]),
    verbose=0
)
print(f"Prediksi setelah load: \n{prediksi}")
```
