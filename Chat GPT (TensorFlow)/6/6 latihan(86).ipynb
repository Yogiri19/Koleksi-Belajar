```python
# Soal Kedua
w = tf.Variable(
    tf.random.normal((1,1))
)
b = tf.Variable(
    tf.random.normal((1,))
)
print(f"w: \n{w.numpy()}")
print(f"b: \n{b.numpy()}")
```

```python
# Soal Ketiga
x = tf.constant([[5.0]])
w = tf.constant([[2.0]])
b = tf.constant([[1.0]])

prediksi = tf.matmul(
    x,
    w
) + b
print(f"prediksi: \n{prediksi}")
```

```python
# Soal Keempat
y_true = tf.constant([
    [3.0],
    [5.0]
])
y_pred = tf.constant([
    [2.0],
    [4.0]
])
loss = tf.reduce_mean(
    (y_true - y_pred) ** 2
)
print(f"loss: \n{loss}")
```

```python
# Soal Kelima
x = tf.constant([[2.0]])
y_true = tf.constant([[5.0]])

w = tf.Variable([[1.0]])
b = tf.Variable([0.0])

with tf.GradientTape() as tape:
    prediksi = tf.matmul(
        x,
        w
    ) + b
    loss = tf.reduce_mean(
        (y_true - y_pred) ** 2
    )
grad_w = tape.gradient(
    loss,
    w
)
print(f"Gradient W: \n{grad_w}")
```

```python
# Soal Keenam
x = tf.constant([[2.0]])
y_true = tf.constant([[5.0]])

w = tf.Variable([[1.0]])
b = tf.Variable([0.0])

with tf.GradientTape() as tape:
    prediksi = tf.matmul(
        x,
        w
    ) + b
    loss = tf.reduce_mean(
        (y_true - y_pred) ** 2
    )
grad_w, grad_b = tape.gradient(
    loss,
    [w,b]
)
print(f"grad_w: \n{grad_w}")
print(f"grad_b: \n{grad_b}")
    
```

```python
# Soal ketujuh
w = tf.Variable([[2.0]])
grad_w = tf.Variable([[0.5]])

lr = 0.1
print(f"Sebelum Update: \n{w.numpy()}")

w.assign_sub(
    lr * grad_w
)
print(f"Sesudah Update: \n{w.numpy()}")
```

```python
# Soal kedelapan
x_train = tf.constant([
    [1.0],
    [2.0]
])
y_train = tf.constant([
    [3.0],
    [5.0]
])

w = tf.Variable([[1.0]])
b = tf.Variable([0.0])

lr = 0.1

with tf.GradientTape() as tape:
    prediksi = tf.matmul(
        x_train,
        w,
    ) + b
    loss = tf.reduce_mean(
        (y_train - prediksi) ** 2
    )
    grad_w,grad_b = tape.gradient(
        loss,
        [w,b]
    )
    w.assign_sub(
        lr * grad_w
    )
    b.assign_sub(
        lr * grad_b
    )
print(f"loss: \n{loss.numpy()}")
print(f"w: \n{w.numpy()}")
print(f"b: \n{b.numpy()}")
```
