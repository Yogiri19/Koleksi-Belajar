```python
import tensorflow as tf

print(f"Tensorflow Version: {tf.__version__}")
```

```python
# Penjumlahan, Pengurangan, Perkalian, dan pembagian
a = tf.constant([10, 20, 30])
b = tf.constant([1, 2, 3])

print(f"a: {a}")
print(f"b: {b}")

print(f"Tambah: {a + b}")
print(f"kurang: {a - b}")
print(f"kali: {a * b}")
print(f"bagi: {a / b}")
```

```python
# Fungsi Matematika TensorFlow
x = tf.constant([4.0, 9.0, 16.0])
print(f"Akar: {tf.sqrt(x)}") # -> Akar 
print(f"Kuadrat: {tf.square(x)}") # -> Kuadrat 2
print(f"Pangkat 3: {tf.pow(x,3)}") # -> Kuadrat 3
```

```python
# Nilai Minuimum dan maksimum
data = tf.constant([10, 50, 20, 90, 15])
print(f"Minimum: {tf.reduce_min(data)}") # -> angka min
print(f"Maksimum: {tf.reduce_max(data)}") # -> angka max
```

```python
# Sum dan Mean
nilai = tf.constant([10, 20, 30, 40, 50])
print(f"Jumlah: {tf.reduce_sum(nilai)}") # Menjumlahkan seluruhnya
print(f"Rata-Rata: {tf.reduce_mean(tf.cast(nilai,tf.float32))}") # cast mengubah int menjadi float 
                                                                 # reduce_mean = menjumlahkan lalu memnagi dengan elementnya
print(f"Ubah data: {tf.cast(nilai,tf.float32)}", ) 
```

```python
# Matrix Addition
a = tf.constant([
    [1, 2],
    [3, 4]
])

b = tf.constant([
    [5, 6],
    [7, 8]
])

hasil = a + b
print(f"Hasil: \n{hasil}")
```

```python
# Matrix Multiplication
a = tf.constant([
    [1, 2],
    [3, 4]
])
b = tf.constant([
    [5, 6],
    [7, 8]
])
hasil = tf.matmul(a,b) # Gak usah diapalin
print(f"hasil: \n{hasil}")
```

```python
# Transpose Matrix
matrix = tf.constant([
    [1, 2, 3],
    [4, 5, 6]
])
print(f"matrix: \n{matrix}")

transpose  = tf.transpose(matrix) # dari baris ke kolom in
print(f"Transpose: \n{transpose}")
```

```python
# Boardcasting dengan scalar
tensor = tf.constant([1, 2, 3, 4])  
hasil = tensor + 10 # Menambah element
print(f"Hasil: \n{hasil}")
```

```python
# Boardcasting dengan Vector
matrix = tf.constant([
    [1, 2, 3],
    [4, 5, 6]
])
vector = tf.constant([10, 20, 30])
hasil = matrix + vector
print(f"Hasil: \n{hasil}")
```

```python
# For Loop pada Tensor
tensor = tf.constant([10, 20, 30, 40, 50])
for i in tensor:
    print(i.numpy())
```

```python
# If else
nilai = tf.constant(85)
if nilai.numpy() >= 75:
    print(f"Lulus")
else:
    print(f"Tidak Lulus")

```

```python
# While Loop
i = tf.Variable(1)
while i.numpy() <= 5:
    print(i.numpy())
    i.assign_add(1)
```

```python
# Studi Kasus Mini
nilai_mahasiswa =  tf.constant([80, 70, 90, 85, 95])
print(f"Jumlah:  \n{tf.reduce_sum(nilai_mahasiswa)}")
print("Rata-Rata:\n",tf.reduce_mean(
    tf.cast(nilai_mahasiswa, tf.float32)
))

print(f"Nilai Tertinggi: \n{tf.reduce_max(nilai_mahasiswa)}")
print(f"Nilai Terendah: \n{tf.reduce_min(nilai_mahasiswa)}")
```
