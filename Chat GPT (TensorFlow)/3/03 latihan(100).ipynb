```python
import tensorflow as tf
print(f"versi: \n{tf.__version__}")
```

```python
a = tf.constant([5, 10, 15])
b = tf.constant([2, 4, 6])

print(f"tambah: \n{a + b}")
print(f"kurang: \n{a - b}")
print(f"kali: \n{a * b}")
print(f"bagi: \n{a / b}")
```

```python
x = tf.constant([1.0, 4.0, 9.0])

print(f"Akar: \n{tf.sqrt(x)}")
print(f"Kuadrat: \n{tf.square(x)}")
print(f"Pangkat 3: \n{tf.pow(x, 3)}")
```

```python
nilai = tf.constant([15, 25, 35, 45, 55])

print(f"Nilai Minimum : \n{tf.reduce_min(nilai)}")
print(f"Nilai Maksimim: \n{tf.reduce_max(nilai)}")
print(f"Jumlah: \n{tf.reduce_sum(nilai)}")
print(f"Rata Rata: \n{tf.reduce_mean(tf.cast(nilai,tf.float32))}")
```

```python
a = tf.constant([
    [2, 4],
    [6, 8]
])

b = tf.constant([
    [1, 3],
    [5, 7]
])

print(f"Penjumlahan: \n{a + b}")
hasil = tf.matmul(a,b)
print(f"Matmul: \n{hasil}")
```

```python
# Soal Kelima
matrix = tf.constant([
    [1, 2, 3],
    [4, 5, 6]
])
vector = tf.constant([10, 20, 30])
trans = tf.transpose(matrix)
print(f"Transpose: \n{trans}")
hasil = matrix + vector
print(f"BoardCasting: \n{hasil}")

data = tf.constant([5, 10, 15, 20, 25])
print(f"Minimum: \n{tf.reduce_min(data)}")
print(f"Maksimum: \n{tf.reduce_max(data)}")
print(f"Jumlah: \n{tf.reduce_sum(data)}")
print(f"rata-rata: \n{tf.reduce_mean(tf.cast(data, tf.float32))}")

for i in data.numpy():
    print(i)
```
