```python
# Soal Pertama
x = tf.Variable(4.0)
with tf.GradientTape() as tape:
    y = x ** 2

grad = tape.gradient(y,x)
print(f"y: \n{y.numpy()}")
print(f"dy/dx: \n{grad.numpy()}")
```

```python
# Soal Kedua
x = tf.Variable(3.0)
with tf.GradientTape() as tape:
    y = x ** 3 + x

grad = tape.gradient(y,x)
print(f"y: \n{y.numpy()}")
print(f"dy/dx: \n{grad.numpy()}")
```

```python
# Soal Ketiga
x = tf.Variable(2.0)
y = tf.Variable(4.0)
with tf.GradientTape() as tape:
    z = x ** 2 + y ** 2

grad = tape.gradient(z,[x,y])
print(f"dz/dx: \n{grad[0].numpy()}")
print(f"dz/dy: \n{grad[1].numpy()}")
```

```python
# Soal Keempat
x = tf.Variable([2.0, 4.0, 6.0])
with tf.GradientTape() as tape:
    y = x ** 2
grad = tape.gradient(y,x)
print(f"dy/dx: \n{grad.numpy()}")
```

```python
# Soal Kelima
x = tf.Variable([1.0, 3.0, 5.0])

with tf.GradientTape() as tape:
    y = tf.reduce_sum(x**2)
grad = tape.gradient(y,x)
print(f"y: \n{y.numpy()}")
print(f"dy/dx:\n{grad.numpy()}")
```

```python
# Soal Keenam
x = tf.Variable(2.0)
with tf.GradientTape(persistent=True) as tape:
    y = x**2
    z = x**3
grad1 = tape.gradient(y,x)
grad2 = tape.gradient(z,x)
print(f"dy/dx: \n{grad1}")
print(f"dz/dx: \n{grad2}")
```

```python
# Soal Kedelapan
import tensorflow as tf
x = tf.Variable(20.0)

with tf.GradientTape() as tape:
    x.assign_sub(
        x - 15
    )

print(x.numpy())
```

```python
# Soal Kesembilan
import tensorflow as tf
x = tf.Variable(0.0)
lr = 0.1

for epoch in range(5):
    with tf.GradientTape() as tape:
        loss = (x-3)**2
    grad = tape.gradient(epoch, loss)
    grad.assign_sub(
        loss - (lr*2)
    )
print(grad)
```

```python
# Soal Kesepuluh
import tensorflow as tf
x = tf.Variable(8.0)
lr = 0.1

for epoch in range(20):
    with tf.GradientTape() as tape:
        loss = x**2 + (2*x) + 1
    grad = tape.gradient(loss, x)
    x.assign_sub(
        lr * grad
    )
print(f"x: \n{x.numpy()}")
print(f"(x): \n{loss.numpy()}")
```
