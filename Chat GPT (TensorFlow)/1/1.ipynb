```python
import tensorflow as tf

print("Tensorflow Version : ", tf.__version__)
```

tf.constant = itu tidak bisa diubah
tf.Variable = itu bisa diubah tapi variable nya harus ditambahkan .numpy()
.assign_add = kelipatan nya
shape 1 dimensi = (element, )
shape 2 dimensi = (Baris, kolom)

```python
# CONSTANT TENSOR
a = tf.constant(10) # Membuat tensor berisi angka 10
b = tf.constant(20)

print(a)
print(b)
```

```python
# OPERASI DASAR
tambah = a + b
kurang = a - b
kali = a * b
bagi = a / b

print("\nTambah: ", tambah.numpy())
print("Kurang: ", kurang.numpy())
print("Kali: ", kali.numpy())
print("Bagi: ", bagi.numpy())
```

```python
# TENSOR VECTOR
vector = tf.constant([10, 20, 30, 40, 50])
print("\nVector")
print(vector)
```

```python
# TENSOR MATRIX
matrix = tf.constant([
    [1, 2, 3 ],
    [4, 5, 6]
])

print("\nMatrix")
print(matrix)
```

```python
# SHAPE
print("\nShape Vector:", vector.shape)
print("Shape Matrix: ", matrix.shape)
```

```python
# VARIABLE
nilai  = tf.Variable(100)
print("\nNilai Awal: ", nilai.numpy())

nilai.assign(200)
print("Nilai Baru: ", nilai.numpy())
```

```python
# LOOP
print("\nLoop Tensor")
for i in vector:
    print(i.numpy())
```

```python
# IF ELSE
angka = tf.constant(80)
if angka.numpy() >= 75:
    print("\nLulus")

else:
    print("\nTidak Lulus")
```

```python
# WHILE
counter = tf.Variable(1)
print("\nWhile Loop")

while counter.numpy() <= 5:
    print(counter.numpy())
    counter.assign_add(1)
```

```python
# LATIHAN OPERASI
x = tf.constant(15)
y = tf.constant(5)

hasil1 = x + y
hasil2 = x - y
hasil3 = x * y
hasil4 = x / y

print("\nHasil Tambah: ", hasil1.numpy())
print("\nHasil kurang: ", hasil2.numpy())
print("\nHasil kali: ", hasil3.numpy())
print("\nHasil bagi: ", hasil4.numpy())
```
